    <b>Codificando la cadena de clase a UTF-8 para eliminar el car√°cter adicional que aparece al cargarlo del ARFF</b>

```python
df[["Class"]]=df[["Class"]].apply(lambda x: x.str.decode('utf8'))
df.sample(10)
```

```python
escalado3 = df.copy()

num_attrs = ['Area','Perimeter','MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity','ConvexArea','EquivDiameter',
            'Extent','Solidity','roundness','Compactness','ShapeFactor1','ShapeFactor3','ShapeFactor4']

scaler = RobustScaler()
escalado3[num_attrs] = scaler.fit_transform(escalado3[num_attrs])

escalado3[num_attrs].describe()
```

<div style="background:#FFFFE0;padding:20px;color:#000000;margin-top:10px;">
    <b>Creaci√≥n de modelos de clasificaci√≥n Bayesianos utilizando una partici√≥n de train-test 70/30.
       Para MinMaxScaler() se utilizara la variable modelo1</b>
</div>

```python
modelo1 = df.copy()

num_attrs = ['Area','Perimeter','MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity','ConvexArea','EquivDiameter',
            'Extent','Solidity','roundness','Compactness','ShapeFactor1','ShapeFactor3','ShapeFactor4']

X = modelo1[num_attrs]
y = modelo1['Class']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

models = []
alphas = [0.1, 0.5, 1.0, 2.0, 5.0]
for alpha in alphas:
    model = GaussianNB(var_smoothing=alpha)
    model.fit(X_train, y_train)
    models.append(model)

for i, model in enumerate(models):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Model {i+1} - Alpha: {alphas[i]}, Accuracy: {acc}")

accuracies = []
for model in models:
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

plt.plot(alphas, accuracies, '-o')
plt.title('Accuracy vs. Alpha')
plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.show()
```

<div style="background:#FFFFE0;padding:20px;color:#000000;margin-top:10px;">
